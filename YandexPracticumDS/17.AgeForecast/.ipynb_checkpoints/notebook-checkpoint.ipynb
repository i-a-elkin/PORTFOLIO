# Определение возраста покупателей

## Исследовательский анализ данных

Импортируем модули.

```python
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense, Conv2D, AveragePooling2D, GlobalAveragePooling2D, Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.applications.resnet import ResNet50
```

Создадим константы.

```python
PATH = "/datasets/faces/"
LABELS = "labels.csv"
FILES = "final_files/"
SEED = 12345
BATCH_SIZE = 32
```

Загрузим датасет.

```python
df = pd.read_csv(PATH + "labels.csv")

print(f"Размер набора данных: {df.shape}")
df.head()
```

Построим распределение возраста.

```python
df["real_age"].hist(bins=100, alpha=0.8, figsize=(10, 6))
plt.vlines(df["real_age"].mean(), 0, 300, color="red", label="mean")
plt.vlines(df["real_age"].median(), 0, 300, color="green", label="median")
plt.title("Распределение возраста")
plt.legend()
plt.show()
```

Создадим загрузчик данных.

```python
def load_train(path):
    train_datagen = ImageDataGenerator(validation_split=0.25, 
                                       rescale=1. / 255,
                                       horizontal_flip=True)
    
    train_gen_flow = train_datagen.flow_from_dataframe(
                                    dataframe=pd.read_csv(path + LABELS),
                                    directory=path + FILES,
                                    x_col='file_name',
                                    y_col='real_age',
                                    target_size=(224, 224),
                                    batch_size=BATCH_SIZE,
                                    class_mode='raw',
                                    seed=SEED,
                                    subset='training')
    
    return train_gen_flow


def load_test(path):
    validation_datagen = ImageDataGenerator(validation_split=0.25, 
                                            rescale=1. / 255)
    
    validation_gen_flow = validation_datagen.flow_from_dataframe(
                                    dataframe=pd.read_csv(path + LABELS),
                                    directory=path + FILES,
                                    x_col='file_name',
                                    y_col='real_age',
                                    target_size=(224, 224),
                                    batch_size=BATCH_SIZE,
                                    class_mode='raw',
                                    seed=SEED,
                                    subset='validation')
    
    return validation_gen_flow
```

Посмотрим сколько файлов содержится в выборках.

```python
train_df = load_train(PATH)
test_df = load_test(PATH)
```

Выведем фотографии на экран.

```python
features, target = next(train_df)

# выводим 16 изображений
fig = plt.figure(figsize=(10,10))
for i in range(16):
    fig.add_subplot(4, 4, i+1)
    plt.imshow(features[i])
    # для компактности удаляем оси и прижимаем изображения друг к другу
    plt.xticks([])
    plt.yticks([])
    plt.tight_layout()
```

Набор данных содержит более 7000 изображений чего должно быть достаточно для обучения сверточной нейросети. Собраные изображения являются портретами людей и могут включать посторонние предметы. Кроме этого изображения могут быть как цветными, так и чернобелыми, могут иметь разное разрешение, разное положение лица в пространстве и разную относительную площадь лица (относительно площади изображения). Можно заметить, что данные ранее были подвергнуты аугментации, т.к. присутствуют повернутые изображения. Обнаруженные особенности не должны значительным образом повлиять на процесс обучения сети, однако для повышения обобщающей способности модели можно провести аугментацию в виде отражения по горизонтали.

Дополнительно можно отметить, что распределение возраста людей близко к нормальному, но имеет положительную скошенность, основной массив изображений собран для людей в возрасте от 15 до 50 лет. Это может повлиять на корректность определения возраста детей и пожилых любей, т.к. они в меньшей степени представлены в наборе данных. Кроме этого на гистограмме можно видеть увеличение количества фотографий для людей юбилейного возраста (кратного десяти), что связано с увеличением фотоматериала во время подобного праздника. Возможно, что при получении низкой величины метрики, значения целевого признака нужно подвергнуть дискретизации или делению на целевые группы для увеличения равномерности распределения.

## Обучение модели

Перенесите сюда код обучения модели и её результат вывода на экран.


(Код в этом разделе запускается в отдельном GPU-тренажёре, поэтому оформлен не как ячейка с кодом, а как код в текстовой ячейке)

```python
def create_model(input_shape):
    # ResNet 50
    optimizer = Adam(lr=0.0005)
    backbone = ResNet50(input_shape=input_shape,
                        weights="/datasets/keras_models/resnet50_weights_tf_dim_ordering_tf_kernels_notop.h5",
                        include_top=False)
    model = Sequential()
    model.add(backbone)
    model.add(GlobalAveragePooling2D())
    model.add(Dense(1, activation="relu"))
    model.compile(optimizer=optimizer, loss="mean_squared_error", metrics=["mean_absolute_error"])
    return model


def train_model(model, train_data, test_data, batch_size=None, epochs=10,
                steps_per_epoch=None, validation_steps=None):
    model.fit(train_data,
              validation_data=test_data,
              batch_size=batch_size, epochs=epochs,
              steps_per_epoch=steps_per_epoch,
              validation_steps=validation_steps,
              verbose=2, shuffle=True)
    return model
```
