#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

## Импорт модулей

```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm

from statsmodels.tsa.seasonal import seasonal_decompose

from sklearn.model_selection import train_test_split, GridSearchCV, TimeSeriesSplit
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, ElasticNet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

## Константы

```python
SCORE = 'neg_root_mean_squared_error'
SCORE_TARGET = 48
SEED = 123
```

## Подготовка

Загрузим данные и выполнить их ресемплирование по одному часу.

```python
df = pd.read_csv('datasets/taxi.csv', parse_dates=[0], index_col=[0])
df.sort_index(inplace=True)
df = df.resample('1H').sum()

display(df.head())
print(f"Дата изменяется монотонно - "
      f"{'ДА' if df.index.is_monotonic else 'НЕТ'}")
print(f"Размер набора данных: {df.shape}")
```

## Анализ

Разложим временной ряд на тренд и сезонную компоненту.

```python
decomposed = seasonal_decompose(df)

plt.figure(figsize=(18, 6))
plt.subplot(131)
decomposed.trend.plot(ax=plt.gca())
plt.title('Тренд')
plt.ylabel('Заказы')
plt.xlabel(None)
plt.subplot(132)
decomposed.seasonal[:48].plot(ax=plt.gca())
plt.title('Сезонность - 48 ч')
plt.xlabel('Дата')
plt.subplot(133)
decomposed.resid.plot(ax=plt.gca())
plt.title('Остатки')
plt.xlabel(None)
plt.tight_layout()
```

Из рисунка видно, что происходит увеличение количества заказов месяц к месяцу. За период с марта по сентябрь количество заказов выросло в 3 раза. Кроме этого можно видеть спад количества заказов в течение суток - оно минимально в утренние часы и максимально в полночь.

## Обучение

### Генерация признаков

Создадим функцию для генерации признаков.

```python
def make_features(data_ts, data_row='num_orders',
                  shifts=(1,),
                  roll_mean_windows=False,
                  roll_std_windows=False,
                  roll_median_windows=False):

	data = data_ts.copy()

	for shift in shifts:
		data[f'shift_{shift}'] = data[data_row].shift(shift)
	if roll_mean_windows:
		for mean_window in roll_mean_windows:
			data[f'roll_mean_{mean_window}'] = data[data_row].shift().rolling(mean_window).mean()
	if roll_std_windows:
		for std_window in roll_std_windows:
			data[f'roll_std_{std_window}'] = data[data_row].shift().rolling(std_window).std()
	if roll_median_windows:
		for median_window in roll_median_windows:
			data[f'roll_median_{median_window}'] = data[data_row].shift().rolling(median_window).median()

	data = data.dropna()
	return data
```

Зададим кросс-валидатор для временных рядов.

```python
tscv = TimeSeriesSplit()
```

Подберем комбинацию признаков, которая позволит получить наименьшую величину метрики.

```python
make_new_features = False
if make_new_features:

	shifts_list = [tuple(range(1, i, 5)) for i in range(4, 56, 5)]
	roll_mean_list = [False, *[tuple(range(4, i, 5)) for i in range(5, 56, 5)]]
	roll_std_list = roll_mean_list.copy()
	roll_median_list = roll_mean_list.copy()

	features_dict = {}
	best_rmse = SCORE_TARGET
	df_train, _ = train_test_split(df, test_size=0.1, shuffle=False)

	for i in tqdm(shifts_list):
		for j in roll_mean_list:
			for k in roll_std_list:
				for l in roll_median_list:

					features = {'shifts': i,
					            'roll_mean_windows': j,
					            'roll_std_windows': k,
					            'roll_median_windows': l}

					df_f = make_features(df_train, **features)
					X_train = df_f.drop(columns=['num_orders'])
					y_train = df_f['num_orders']

					# параметры линейной регрессии по умолчанию
					pipe_f_lr = Pipeline([('scaler', StandardScaler()),
					                      ('lr', LinearRegression())])

					params_f_lr = {'lr__fit_intercept': [True]}

					search_f_lr = GridSearchCV(estimator=pipe_f_lr,
					                           param_grid=params_f_lr,
					                           scoring=SCORE,
					                           cv=tscv,
					                           n_jobs=-1,
					                           verbose=0)

					search_f_lr.fit(X_train, y_train)

					if best_rmse > - search_f_lr.best_score_:
						best_rmse = - search_f_lr.best_score_
						features_dict = features

	print(f'Лучшее RMSE: {best_rmse:.2f}')
	print(f'Признаки: {features_dict}')
```

Лучшее RMSE: 26.82
Признаки: {'shifts': (1, 6, 11, 16, 21, 26, 31, 36, 41, 46, 51),
'roll_mean_windows': (4, 9, 14, 19, 24, 29, 34, 39, 44, 49),
'roll_std_windows': (4, 9, 14, 19, 24, 29, 34, 39, 44, 49),
'roll_median_windows': (4, 9, 14)}

Сгенерируем новые признаки и разделим выборку на тренировочную и тестовую части.

```python
features = {'shifts': (1, 6, 11, 16, 21, 26, 31, 36, 41, 46, 51),
            'roll_mean_windows': (4, 9, 14, 19, 24, 29, 34, 39, 44, 49),
            'roll_std_windows': (4, 9, 14, 19, 24, 29, 34, 39, 44, 49),
            'roll_median_windows': (4, 9, 14)}

df_f = make_features(df, **features)
X = df_f.drop(columns=['num_orders'])
y = df_f['num_orders']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, shuffle=False)

display(X_train.head())
print(f"Тренировочная выборка монотонна - {'ДА' if X_train.index.is_monotonic else 'НЕТ'}\n"
	  f"Тестовая выборка монотонна - {'ДА' if X_test.index.is_monotonic else 'НЕТ'}")
print(f"Размер тренировочной выборки: {X_train.shape}\n"
	  f"Размер тестовой выборки: {X_test.shape}")
```

### Линейная регрессия

Подберем гиперпараметры модели.

```python
search_hp_lr = False
if search_hp_lr:
	pipe_lr = Pipeline([('scaler', StandardScaler()),
						('lr', ElasticNet(max_iter=10000))])

	params_lr = {'lr__l1_ratio': [0, 1],
				 'lr__alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000],
				 'lr__fit_intercept': [True, False]}

	grid_lr = GridSearchCV(estimator=pipe_lr,
						   param_grid=params_lr,
				  		   scoring=SCORE,
				  		   cv=tscv,
						   n_jobs=-1,
						   verbose=1)

	grid_lr.fit(X_train, y_train)
	print(grid_lr.best_params_)
```

{'lr__alpha': 0.01, 'lr__fit_intercept': True, 'lr__l1_ratio': 1}

Обучим модель с найденными гиперпараметрами.

```python
pipe_lr = Pipeline([('scaler', StandardScaler()),
					('lr', ElasticNet(max_iter=10000))])

params_lr = {'lr__l1_ratio': [1],
			 'lr__alpha': [0.01],
			 'lr__fit_intercept': [True]}

grid_lr = GridSearchCV(estimator=pipe_lr,
					   param_grid=params_lr,
					   scoring=SCORE,
					   cv=tscv,
					   n_jobs=-1,
					   verbose=1)

grid_lr.fit(X_train, y_train)
print(f'RMSE линейной регрессии: {- grid_lr.best_score_:.2f}')
```

### Случайный лес

Подберем гиперпараметры модели.

```python
search_hp_rf = False
if search_hp_rf:
	params_rf = {'n_estimators': [50, 100, 200, 300, 400, 500],
				 'max_depth': [1, 2, 4, 6, 8, 10, 12, 15, 20]}

	grid_rf = GridSearchCV(estimator=RandomForestRegressor(random_state=SEED),
						   param_grid=params_rf,
						   scoring=SCORE,
						   cv=tscv,
						   n_jobs=-1,
						   verbose=1)

	grid_rf.fit(X_train, y_train)
	print(grid_rf.best_params_)
```

{'max_depth': 20, 'n_estimators': 500}

Обучим модель с найденными гиперпараметрами.

```python
params_rf = {'n_estimators': [500],
			 'max_depth': [20]}

grid_rf = GridSearchCV(estimator=RandomForestRegressor(random_state=SEED),
					   param_grid=params_rf,
					   scoring=SCORE,
					   cv=tscv,
					   n_jobs=-1,
					   verbose=1)

grid_rf.fit(X_train, y_train)
print(f'RMSE случайного леса: {- grid_rf.best_score_:.2f}')
```

### Градиентный бустинг

Подберем гиперпараметры модели.

```python
search_hp_gb = False
if search_hp_gb:
	params_gb = {'n_estimators': [50, 100, 200, 300, 400, 500],
				 'max_depth': [1, 2, 4, 6, 8, 10, 12, 15, 20],
				 'learning_rate': [0.0001, 0.001, 0.01, 0.1, 0.5, 1]}

	grid_gb = GridSearchCV(estimator=GradientBoostingRegressor(random_state=SEED),
						   param_grid=params_gb,
						   scoring=SCORE,
						   cv=tscv,
						   n_jobs=-1,
						   verbose=1)

	grid_gb.fit(X_train, y_train)
	print(grid_gb.best_params_)
```

{'learning_rate': 0.01, 'max_depth': 4, 'n_estimators': 500}

Обучим модель с найденными параметрами.

```python
params_gb = {'n_estimators': [500],
			 'max_depth': [4],
			 'learning_rate': [0.01]}

grid_gb = GridSearchCV(estimator=GradientBoostingRegressor(random_state=SEED),
					   param_grid=params_gb,
					   scoring=SCORE,
					   cv=tscv,
					   n_jobs=-1,
					   verbose=1)

grid_gb.fit(X_train, y_train)
print(f'RMSE градиентного бустинга: {- grid_gb.best_score_:.2f}')
```

Объединим результаты в таблицу и построим диаграмму.

```python
rmse_scores = pd.DataFrame(data={'rmse': [- grid_lr.best_score_, - grid_rf.best_score_, - grid_gb.best_score_]},
						   index=pd.Index(data=['lr', 'rf', 'gb'], name='models')).round(2).sort_values('rmse')

rmse_scores.plot(kind='bar', rot=0, figsize=(6,5), title='Метрики моделей')
plt.show()
print(rmse_scores.transpose())
```

Наименьшая величина метрики достигнута с помощью линейной регрессии.

## Тестирование

Проверим модель на тестовой выборке.

```python
predict = grid_lr.predict(X_test)

plt.figure(figsize=(12, 6))
plt.scatter(y_test.index, y_test.values, alpha=0.5, label='true_values')
plt.plot(y_test.index, predict, color='red', label='predict_values')
plt.gcf().autofmt_xdate()
plt.ylabel('Заказы')
plt.xlabel('Дата')
plt.title('Предсказание на тестовой выборке')
plt.legend()
plt.show()

print(f'RMSE на тестовой выборке: '
	  f'{mean_squared_error(y_test, predict, squared=False):.2f}')
```

## Вывод

С помощью полученной модели на тестовой выборке удалось добиться метрики RMSE = 43, что удовлетворяет условию задачи. Основным фактором, оказывающим негативное влияние на искомую метрику является наличие редких краткосрочных событий с повышенным или наоборот пониженным спросом на такси.
