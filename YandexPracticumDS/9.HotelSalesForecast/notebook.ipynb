# Прогнозирование оттока клиентов в сети отелей «Как в гостях»

## Описание проекта
Заказчик этого исследования — сеть отелей «Как в гостях».

Чтобы привлечь клиентов, эта сеть отелей добавила на свой сайт возможность забронировать номер без предоплаты. Однако если клиент отменял бронирование, то компания терпела убытки. Сотрудники отеля могли, например, закупить продукты к приезду гостя или просто не успеть найти другого клиента.

Чтобы решить эту проблему, вам нужно разработать систему, которая предсказывает отказ от брони. Если модель покажет, что бронь будет отменена, то клиенту предлагается внести депозит. Размер депозита — 80% от стоимости номера за одни сутки и затрат на разовую уборку. Деньги будут списаны со счёта клиента, если он всё же отменит бронь.

## Бизнес-метрика и другие данные
Основная бизнес-метрика для любой сети отелей — её прибыль. Прибыль отеля — это разница между стоимостью номера за все ночи и затраты на обслуживание: как при подготовке номера, так и при проживании постояльца.

В отеле есть несколько типов номеров. В зависимости от типа номера назначается стоимость за одну ночь. Есть также затраты на уборку. Если клиент снял номер надолго, то убираются каждые два дня.

Стоимость номеров отеля:
- категория A: за ночь — 1 000, разовое обслуживание — 400;
- категория B: за ночь — 800, разовое обслуживание — 350;
- категория C: за ночь — 600, разовое обслуживание — 350;
- категория D: за ночь — 550, разовое обслуживание — 150;
- категория E: за ночь — 500, разовое обслуживание — 150;
- категория F: за ночь — 450, разовое обслуживание — 150;
- категория G: за ночь — 350, разовое обслуживание — 150.

В ценовой политике отеля используются сезонные коэффициенты: весной и осенью цены повышаются на 20%, летом — на 40%.

На разработку системы прогнозирования заложен бюджет — 400 000. При этом необходимо учесть, что внедрение модели должно окупиться за год. Затраты на разработку должны быть меньше той выручки, которую система принесёт компании.

## Формулировка ML-задачи на основе бизнес-задачи

Соберем данные для расчета прибыли в словари.

```python
NIGHT = {'A': 1000, 'B': 800, 'C': 600, 'D': 550, 'E': 500, 'F': 450, 'G': 350}
SERVICE = {'A': 400, 'B': 350, 'C': 350, 'D': 150, 'E': 150, 'F': 150, 'G': 150}
COEFFICIENT = {'December': 1, 'January': 1, 'February': 1, 
               'March': 1.2, 'April': 1.2, 'May': 1.2,
               'June': 1.4, 'July': 1.4, 'August': 1.4,
               'September': 1.2, 'October': 1.2, 'November': 1.2}
DEPOSIT = 0.8
COST = 400000
```

Напишем функцию для расчета прибыли отеля до введения депозитов. Расчет будем производить по следущей схеме:
- количество ночей умножим на цену за 1 ночь, с учетом котегории номера
- умножим полученное значение на сезонный коэффициент
- умножим на ноль, если клиент отменил бронирование или на 1, если прибыл
- вычтем затраты на уборку, с учетом категории номера (по условию задачи номер убирают 1 раз в два дня, т.е. 1 ночь - 1 раз, 2 ночи - 1 раз, 3 ночи - 2 раза, 4 ночи - 2 раза, 5 ночей - 3 раза и т.д.)
- затраты за первую ночь отель несет даже если клиент не приехал (по условию задачи), за оставшиеся, если приехал, согласно условию проживания
- симмируем полученные значения

```python
def profit_before_deposit(data_ohe: pd.DataFrame, is_canceled_true: pd.Series):
    """
    Функция принимает на фход датафрейм, в котором категориальные признаки
    закодированы методом OHE и серию с фактом отмены бронирования.
    Выбирает столбцы, необходимые для расчетов.
    Расчитывает прибыль на 1 клиента с учетом количества ночей, 
    категории номера, сезона и расходов на уборку периодичностью 1 раз в 2 дня.
    Расходы на уборку в первую ночь отель несет даже в случае отмены бронирования.
    Выводит суммарную прибыль от всех клиентов.
    """
    nights = ['stays_in_weekend_nights', 'stays_in_week_nights']
    room = 'reserved_room_type'
    month = 'arrival_date_month'
    
    df_pay_per_night = pd.DataFrame()
    df_service = pd.DataFrame()
    df_coefficient = pd.DataFrame()
    
    for col in data_ohe.columns:
        if room in col:
            df_pay_per_night[col] = data_ohe[col] * NIGHT[col.split('_')[-1]]
            df_service[col] = data_ohe[col] * SERVICE[col.split('_')[-1]]
        elif month in col:
            df_coefficient[col] = data_ohe[col] * COEFFICIENT[col.split('_')[-1]]
    
    pay_per_night = df_pay_per_night.sum(axis=1)
    service = df_service.sum(axis=1)
    coefficient = df_coefficient.sum(axis=1)
    nights_num = data_ohe[nights].sum(axis=1)
    
    revenue_per_client = ((1 - is_canceled_true) * 
                          pay_per_night * 
                          nights_num * 
                          coefficient)
    
    expenses_per_client = (service + 
                           (1 - is_canceled_true) * 
                           service * ((nights_num + 1) // 2 - 1))
        
    profit_per_client = revenue_per_client - expenses_per_client
    profit = profit_per_client.sum()
    
    return profit
```

Расчитаем прибыль отеля по тренировочной выборке за 2015 и 2016 гг. и за 2017 г. по тестовой.

Напишем функцию для расчета прибыли после введения депозитов. Расчет будем проводить по следующей схеме:
- если предсказание совпадает с реальным отказом (1 и 1), то прибыль отеля составляет размер депозита минус затраты на сервис
- если модель предсказала отсутствие отказа, а клиент отказался (0 и 1), то отель терпит убыток в сумме затрат на сервис
- в остальных случаях (1 и 0; 0 и 0) расчет прибыли осуществляем также, как и до введения депозитов

```python
def profit_after_deposit(data_ohe: pd.DataFrame, 
                         is_canceled_true: pd.Series, 
                         is_canceled_predict: pd.Series,
                         coeff_deposit: float):
    """
    На вход поступает датафрейм с исходными данными закодированными OHE 
    без масштабирования, серия с истинными значениями и серия 
    с предсказанными значениями отказов, а также коэффициент депозита.
    Датафрейм и серии должны иметь одинаковые индексы.
    В датафрейме выбираются нужные колонки для расчетов, для каждого клиента
    вычисляются цена за ночь с учетом категории номера и сезонного коэффициента,
    цена обслуживания с учетом категории номера, количество ночей.
    После этого вычисляется прибыль, получаемая с каждого клиента, 
    с учетом истинных и предсказанных значений отказа от бронирования.
    На выход поступает суммарная прибыль от всех клиентов.
    """
    nights = ['stays_in_weekend_nights', 'stays_in_week_nights']
    room = 'reserved_room_type'
    month = 'arrival_date_month'
    
    df_pay_per_night = pd.DataFrame()
    df_service = pd.DataFrame()
    df_coefficient = pd.DataFrame()
    df_profit = pd.DataFrame()
       
    for col in data_ohe.columns:
        if room in col:
            df_pay_per_night[col] = data_ohe[col] * NIGHT[col.split('_')[-1]]
            df_service[col] = data_ohe[col] * SERVICE[col.split('_')[-1]]
        elif month in col:
            df_coefficient[col] = data_ohe[col] * COEFFICIENT[col.split('_')[-1]]
    
    df_profit['pay_per_night'] = df_pay_per_night.sum(axis=1)
    df_profit['service'] = df_service.sum(axis=1)
    df_profit['coefficient'] = df_coefficient.sum(axis=1)
    df_profit['nights_num'] = data_ohe[nights].sum(axis=1)
    df_profit['is_canceled_true'] = is_canceled_true
    df_profit['is_canceled_predict'] = is_canceled_predict
    
    for index, row in df_profit.iterrows():
        if (row['is_canceled_true'] == 1) and (row['is_canceled_predict'] == 1):
            deposit = coeff_deposit * (row['pay_per_night'] * row['coefficient'] + row['service'])
            df_profit.loc[index, 'profit_per_client'] = deposit
        elif (row['is_canceled_true'] == 1) and (row['is_canceled_predict'] == 0):
            service = row['service']
            df_profit.loc[index, 'profit_per_client'] = 0 - service
        else:
            revenue_per_client = (row['pay_per_night'] * 
                                  row['nights_num'] * 
                                  row['coefficient'])
            expenses_per_client = (row['service'] * 
                                  ((row['nights_num'] + 1) // 2))
            profit_per_client = revenue_per_client - expenses_per_client
            df_profit.loc[index, 'profit_per_client'] = profit_per_client
    
    profit = df_profit['profit_per_client'].sum()
    return profit
```

## Портрет «ненадёжного» клиента

Для того чтобы выявить склонность клиента к отказу, выведем на экран коэффициенты линейной модели (за исключением нулевых). Положительные коэффициенты увеличивают вероятность отказа, а отрицательные уменьшают.

```python
coef_logreg = pd.DataFrame(best_logreg.coef_[0], 
                           index=X_test.columns, columns=['coef'])
coef_logreg['coef_abs'] = (coef_logreg['coef']).abs()
coef_logreg = coef_logreg[coef_logreg['coef'] != 0]\
              .sort_values('coef_abs')['coef']

%matplotlib inline
plt.style.use('ggplot')
coef_logreg.plot(kind='barh', fontsize=13, figsize=(8, 9))
plt.title('Logistic Regression Classifier', 
          fontsize=16)
plt.xlabel('Coefficient Value', fontsize=14)
plt.show()
```

Можно видеть, что наибольшее влияние на отказ от бронирование оказывает количество ранее отменённых заказов и количество дней между датой бронирования и датой прибытия. Чем больше величина этих признаков тем ненадежнее клиент. 

Надежность клиента увеличивается от следующих факторов (чем больше тем надежнее клиент):
- тип заказчика не связан с договором или групповым заездом, но связан с бронированием типа Transient
- необходимы места для парковки
- в заказ включен завтрак
- в параметры заказа вносились изменения
- канал дистрибуции TA/TO или Direct
- клиент имеет подтверждённые ранее заказы 

Остальные признаки имеют несущественное влияние на вероятность отказа. Результаты исследовательского анализа данных, проведенного с помощью sweetviz согласуются с найденными закономерностями.
